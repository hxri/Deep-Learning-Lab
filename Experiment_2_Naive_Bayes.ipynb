```python
X = data['data']
y = data['target']
```

```python
classifiers = [GaussianNB(), BernoulliNB(binarize = 5.0), MultinomialNB()]
```

## Wine Dataset

```python
def train(classifiers, X, y, split):

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split, random_state=42)

  print("Data dimensions\n")
  print("Train Data :", str(X_train.shape) + " and " + str(y_train.shape))
  print("Test Data :", str(X_test.shape) + " and " + str(y_test.shape))
  print("--------------------\n\n")
  
  for i in range(len(classifiers)):
    print(str(classifiers[i]))
    print("\n")
    classifiers[i].fit(X_train, y_train)
    y_pred = classifiers[i].predict(X_test)
    print(y_pred)
    print("\n")
    print(classification_report(y_test, y_pred))
    print("--------------------\n\n")
```

```python
train(classifiers, X, y, 0.10)
```
